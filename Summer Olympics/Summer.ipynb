# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
from matplotlib import * 
import pandas as pd
df = pd.read_csv("summer.csv")

cities = df["City"].unique()
print(len(cities))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df1 = df[df["Medal"] == "Gold"]
sports = df["Sport"].unique()
data = []
for sport in sports:
    data.append([sport, len(df1[df1["Sport"]==sport])])
data = pd.DataFrame(data, columns =['sport', 'gold_medals'])
data = data.sort_values(by='gold_medals',ascending= False).head()
print(data)
data.plot( x = "sport", y= "gold_medals", kind = "bar")
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
sports = df["Sport"].unique()
data = []
for sport in sports:
    data.append([sport, len(df[df["Sport"]==sport])])
data = pd.DataFrame(data, columns =['sport', 'medals'])
data = data.sort_values(by='medals',ascending= False).head()
print(data)
data.plot( x = "sport", y= "medals", kind = "bar")
```

### 4. Which player has won most number of medals? (Top 5)

```python
players = df["Athlete"].unique()
data = []
for player in players:
    data.append([player, len(df[df["Athlete"]==player])])
data = pd.DataFrame(data, columns =['player', 'medals'])
data = data.sort_values(by='medals',ascending= False).head()
print(data)
data.plot( x = "player", y= "medals", kind = "bar")
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
df1 = df[df["Medal"] == "Gold"]
players = df["Athlete"].unique()
data = []
for player in players:
    data.append([player, len(df1[df1["Athlete"]==player])])
data = pd.DataFrame(data, columns =['player', 'gold_medals'])
data = data.sort_values(by='gold_medals',ascending= False).head()
print(data)
data.plot( x = "player", y= "gold_medals", kind = "bar")
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python

df1 = df[df["Medal"] == "Gold"]
df2 = df1[df1["Country"] == "IND"]
years = df2['Year'].unique()
data = []
for year in years :
    data.append([year, len(df2[df2["Year"]== year])])
data = pd.DataFrame(data, columns =['year', 'gold_medals'])
data = data.sort_values(by='year',ascending= True).head(10)
print(data)

```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
events = df["Event"].unique()
data = []
for event in events :
    data.append([event, len(df[df["Event"]== event])])
data = pd.DataFrame(data, columns =['event', 'freq'])
data = data.sort_values(by='freq',ascending= False).head()
print(data)
data.plot( x = "event", y= "freq", kind = "bar")
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df1 = df[df["Medal"] == "Gold"]
df2 = df1[df1["Gender"]== "Women"]
sports = df2["Sport"].unique()
data = []
for sport in sports:
    data.append([sport, len(df2[df2["Sport"]== sport])])
data = pd.DataFrame(data, columns =['sport', 'freq'])
data = data.sort_values(by='freq',ascending= False).head()
print(data)
data.plot( x = "sport", y= "freq", kind = "bar")
```
